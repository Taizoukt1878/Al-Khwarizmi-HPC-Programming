## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing Files

```python
data_O0 = pd.read_csv('matrice_o0.txt')
data_O2 = pd.read_csv('matrice_o2.txt')
```

```python
strides = data_O0['stride'].values
rates = data_O0['rate'].values
msecs = data_O0['msecs'].values
##########
strides2 = data_O2['stride'].values
rates2 = data_O2['rate'].values
msecs2 = data_O2['msecs'].values
```

## Visualisation

```python
plt.plot(strides,rates,'r', label = 'O0')
plt.plot(strides2,rates2,'b', label = 'O2')
plt.title('Variation de la bande passante / stride')
plt.xlabel('Stride')
plt.ylabel('rate')
plt.grid()
plt.legend()
```

```python
plt.plot(strides,msecs,'r', label = 'O0')
plt.plot(strides2,msecs2,'b', label = 'O2')
plt.title('Variation du cpu time / stride')
plt.xlabel('Stride')
plt.ylabel('cpu_time')
plt.grid()
plt.legend()
```
